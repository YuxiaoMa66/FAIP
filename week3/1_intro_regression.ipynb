**Slicing Arrays**

Sometimes we want just to check one specific value from a given array, and it can be done by referencing its position
within the bigger array (taking into account that position within the array start at position `0`).

Say we want to select the element from the second row and third column from our previous variable `array2d`.

First take a look at `array2d`

````python
# Show original array
array2d
````

The element from the second row and third column would be the `33`.

To select it we just select its position within the array (start counting from 0).

````python
array2d[1,2]
````

But some other times, you may want to work only with certain parts of an array. We can use the called array slicing
for this purpose. It follows a similar notation as selecting single values but this time we are going to select
the desired intervals `(start:end)`. If we want to select a the whole size of one of the dimensions, we can use `:`.

Look at the following example in which we want to select a sub-matrix from our `array2d` taking all rows (`:`) but only the second and third column (`1:3`).

````python
# We want to select a submatrix:
array2d[:, 1:3]
````

It also works the other way round. We can select all columns but only certain rows.

````python
# We want to select just the second row of array2d:
array2d[1, :]
````

We can also use this notation to set the value of multiple entries within the selected interval. The following example takes the numbers from the third position until the fifth and replaces those values with `13`.

````python
array5 = np.arange(0, 10)
print(array5)
array5[2:5] = 13
print(array5)
````

**Exercise 5** Building machine learning models with Scikit

````python
from sklearn.neighbors import NearestCentroid
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
np.random.seed(42)

u = NearestCentroid()
````

`u` is defined as an untrained nearest centroid classifier. It is untrained because we did not specify any data, `u` therefore only �remembers� that in the future we may want to apply the nearest centroid classifier to some data.

Below, we load the breast cancer data, again using only the first 10 features.

````python
breast_cancer = load_breast_cancer()

X = breast_cancer.data[:, :10] # Features
y = breast_cancer.target # Labels
````

Let split our dataset into 2 parts, the training set and the test set. We want to assign 50% of our available samples to the test set. That way we are able to fit (or train) our classifier on half of our data, leaving the other half to be used for evaluation.

````python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.5)

u.fit(X_train, y_train)
````

Now it can by applied to the 10-dimensional input (data points) and classify them as either of the classes: �healthy� (0) and �sick� (1). In order to calculate the predictions we use the `predict()` function. 

````python
test_predictions = u.predict(X_test)
test_predictions
````

**Exercise 6**  Plotting classifiers.

When working with 2-dimensional datasets, we can visualize the decision boundary of our classifier. For this purpose we will use the following helper functions.

````python
def plot_dataset(X, y):
    plt.plot(X[y==0, 0], X[y==0, 1], "bs")
    plt.plot(X[y==1, 0], X[y==1, 1], "g^")

    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')

def plot_decision_boundary(X, y, model):
    Xmin = np.min(X[:,:],axis=0)
    Xmax = np.max(X[:,:],axis=0)

    x0, x1 = np.meshgrid(
            np.linspace(Xmin[0], Xmax[0], 500).reshape(-1, 1),
            np.linspace(Xmin[1], Xmax[1], 200).reshape(-1, 1),
        )
    X_new = np.c_[x0.ravel(), x1.ravel()]

    y_new = model.predict(X_new)

    plot_dataset(X,y)

    zz = y_new.reshape(x0.shape)
    contour = plt.contour(x0, x1, zz, levels=np.array([0.5]),colors='k')
````

Now let us use the first 50 points from the `breast_cancer` dataset and two of its features.

````python
clf = NearestCentroid()

clf.fit(X, y)

plot_decision_boundary(X, y, clf)
plt.show()
````

If these new points were part of the training set, and we used k=5, the predictions would be calculated based on the average of the five nearest training points. If the x-values ​​of the new points were much larger than 3.25, the nearest five neighbors of these new points might change (depending on the distribution of these new points), possibly resulting in a different average y-value.
If these new points were part of the test set, the predictions would be based only on the average of the y-values ​​of the original five points (if we had no more training data).
