### Query all flights from June 15th

```python
# Query all flights from June 15th
# convert the timezone to None of all datetime columns: fobt, aobt, fat, aat
flights['fobt'] = flights['fobt'].dt.tz_convert(None)
flights['aobt'] = flights['aobt'].dt.tz_convert(None)
flights['fat'] = flights['fat'].dt.tz_convert(None)
flights['aat'] = flights['aat'].dt.tz_convert(None)


f15 = flights.query("fobt >= '2020-06-15' and fobt < '2020-06-16'")
f15
```

```python
# function to calculate great circle distance between two locations
def calc_gcd(r):
    lat1 = np.radians(r.lat_orig)
    lon1 = np.radians(r.lon_orig)
    lat2 = np.radians(r.lat_dest)
    lon2 = np.radians(r.lon_dest)

    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    c = 2 * np.arcsin(np.sqrt(a))
    d = c * 6371 * 0.539957  # nautical miles
    return d
```

## Query flights from June 15th, and then calculate the great circle distance with apply()

```python
# Query flights from June 15th, and then calculate the great circle distance with apply()
gcd15 = f15.assign(gcd=lambda d: d.apply(calc_gcd, axis=1)).head()
gcd15
```

### With one nest command: 
- add a new column named **gcd** (great circle distance) for all the flights
- then add another new column that calculates the different between **distance** and **gcd**, with column name **distance_diff**

```python
# add "gcd" column to the dataframe
flights = flights.assign(gcd=lambda d: d.apply(calc_gcd, axis=1))

# add "distance_diff" column to the dataframe
flights = flights.assign(distance_diff=lambda d: d.distance - d.gcd)

flights.head()
```

### Nesting operations

- Query flights originated from Paris Charles de Gaulle Airport (airport code: LFPG)
- Calculate arrival delays of all airports
- Group by destination
- Calculated the mean flight distance and average arrival delay at the destination airports
- Sow and sort results for destinations with average arrival delay large than 10 minutes with more than 5 flights

```python

# 1
flights_from_lfpg = flights.query("origin == 'LFPG'")

# 2
flights_from_lfpg_delay = flights_from_lfpg.assign(arrival_delay=(flights_from_lfpg.aat - flights_from_lfpg.fat).dt.total_seconds() / 60)   # in minutes cuz the last question is about minutes

# 3: group by destination 
destinations = flights_from_lfpg_delay.groupby('destination')
# # check the destinations' first group
# destinations.get_group('EGLL')
# calculate the number of flights for each destination
num_flights = destinations.size()

# 4: add mean distance and mean arrival delay to destinations
destinations = destinations.agg({'distance': 'mean', 'arrival_delay': 'mean'})

# 5: arrival delay large than 10 minutes 
# asked_destinations = destinations.query("arrival_delay > 10")
# with more than 5 flights according to the num_flights
# asked_destinations = asked_destinations[num_flights > 5]
destinations["num_flights"] = num_flights
asked_destinations = destinations.query("arrival_delay > 10 and num_flights > 5")

asked_destinations = asked_destinations.sort_values('arrival_delay', ascending=False)

asked_destinations

# flights_from_lfpg
















```

### Plot the departure and arrival delay trends over the 24-hours for at Paris Charles de Gaulle Airport (LFPG) airport
-  based on all flights departing and arriving at LFPG
- you should consider the variations for different days

```python
arrival_delay = (flights['aat'] - flights['fat']).dt.total_seconds() / 60
departure_delay = (flights['aobt'] - flights['fobt']).dt.total_seconds() / 60

flights['arrival_delay'] = arrival_delay
flights['departure_delay'] = departure_delay

# Filter out flights departing from and arriving at LFPG
lfpg_departures = flights[flights['origin'] == 'LFPG'].copy(deep=True)
lfpg_arrivals = flights[flights['destination'] == 'LFPG'].copy(deep=True)

# Create new columns to extract the hour from fobt and aat
lfpg_departures['departure_hour'] = lfpg_departures['fobt'].dt.hour  # Extract the hour of departure
lfpg_arrivals['arrival_hour'] = lfpg_arrivals['aat'].dt.hour  # Extract the hour of arrival

# Group by hour and calculate the average departure delay and arrival delay for each hour
departure_delays = lfpg_departures.groupby('departure_hour')['departure_delay'].mean()
arrival_delays = lfpg_arrivals.groupby('arrival_hour')['arrival_delay'].mean()

# Plot the trends of departure and arrival delays over 24 hours
plt.figure(figsize=(12, 6))

# Plot departure delays (flights departing from LFPG)
plt.plot(departure_delays.index, departure_delays, label='Departure Delay (LFPG)')

# Plot arrival delays (flights arriving at LFPG)
plt.plot(arrival_delays.index, arrival_delays, label='Arrival Delay (LFPG)', linestyle='--')


plt.title('Departure and Arrival Delay Trends over 24 Hours at LFPG')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Delay (minutes)')
plt.legend()
plt.grid(True)

plt.show()
```

### Use jointplot() to show the relationship between departure and arrival delays with 2D kde plot, you should

 - randomly sample 10000 flights for the plot for better performance
 - limit showing delays between -60 and 60 minutes for better visualization
 

```python
# Randomly sample 10,000 flights
sampled_flights = flights.sample(n=10000, random_state=66)

# Filter delays between -60 and 60 minutes for better visualization
filtered_flights = sampled_flights[(sampled_flights['departure_delay'] >= -60) & (sampled_flights['departure_delay'] <= 60) &
                                   (sampled_flights['arrival_delay'] >= -60) & (sampled_flights['arrival_delay'] <= 60)]

# sns.jointplot to show the relationship with a 2D KDE plot
sns.jointplot(x='departure_delay', y='arrival_delay', data=filtered_flights, kind='kde', fill=True)

plt.xlim(-60, 60)
plt.ylim(-60, 60)

plt.suptitle('Relationship between Departure and Arrival Delays (KDE Plot)')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')

# Show plot
plt.show()








```
