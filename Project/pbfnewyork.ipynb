```python
import osmium as osm
from collections import defaultdict

# 自定义处理类，用于统计标签的种类
class TagInspector(osm.SimpleHandler):
    def __init__(self):
        osm.SimpleHandler.__init__(self)
        # 用于存储不同元素类型中的标签统计
        self.tag_stats = {
            'nodes': defaultdict(int),
            'ways': defaultdict(int),
            'relations': defaultdict(int)
        }

    def node(self, n):
        # 遍历每个节点的标签，并统计标签名称出现的次数
        for tag in n.tags:
            self.tag_stats['nodes'][tag.k] += 1

    def way(self, w):
        # 遍历每个道路元素的标签，并统计标签名称出现的次数
        for tag in w.tags:
            self.tag_stats['ways'][tag.k] += 1

    def relation(self, r):
        # 遍历每个关系的标签，并统计标签名称出现的次数
        for tag in r.tags:
            self.tag_stats['relations'][tag.k] += 1

# 实例化处理类
inspector = TagInspector()

# 读取 .osm.pbf 文件
inspector.apply_file("new-york-latest.osm.pbf")

# 打印统计结果
print("Node Tags:")
for tag, count in inspector.tag_stats['nodes'].items():
    print(f"{tag}: {count} occurrences")

print("\nWay Tags:")
for tag, count in inspector.tag_stats['ways'].items():
    print(f"{tag}: {count} occurrences")

print("\nRelation Tags:")
for tag, count in inspector.tag_stats['relations'].items():
    print(f"{tag}: {count} occurrences")
```

```python
import osmnx as ox
import geopandas as gpd
from tqdm import tqdm
import gc

# 纽约市的经纬度范围
north, south, east, west = 40.917577, 40.477399, -73.700272, -74.259090

# 分块的数量（3x3 共9块）
num_chunks = 3

# 计算每块的经纬度范围
lat_step = (north - south) / num_chunks
lon_step = (east - west) / num_chunks

last_amenities_filtered_chunk = None

# 逐块读取和保存数据
for i in tqdm(range(num_chunks), desc="Latitude chunks"):
    for j in range(num_chunks):
        # 计算每块的边界
        north_bound = north - i * lat_step
        south_bound = north_bound - lat_step
        west_bound = west + j * lon_step
        east_bound = west_bound + lon_step

        # 使用 OSMNX 读取特定范围内的设施数据
        bbox = (south_bound, north_bound, west_bound, east_bound)
        amenities = ox.geometries_from_bbox(*bbox, tags={'amenity': True})

        # 选择有用的列
        useful_amenity_columns = ['amenity', 'name', 'geometry']
        amenities_filtered_chunk = amenities[useful_amenity_columns]

        # 保存每个块为 GeoJSON 文件
        filename = f'filtered_amenities_part_{i + 1}_{j + 1}.geojson'
        amenities_filtered_chunk.to_file(filename, driver='GeoJSON')

        # 打印保存信息
        print(f"Saved chunk {i + 1}_{j + 1} to {filename}")
        print(amenities_filtered_chunk.head())

        # 释放内存
        del amenities
        gc.collect()

        # 只保留最后一个块的数据
        last_amenities_filtered_chunk = amenities_filtered_chunk

# 最后一个块的数据在这里可用
print("Last amenities chunk:")
print(last_amenities_filtered_chunk.head())
```

```python
import osmnx as ox
import geopandas as gpd
from tqdm import tqdm
import gc

# 纽约市的经纬度范围
north, south, east, west = 40.917577, 40.477399, -73.700272, -74.259090

# 分块的数量（3x3 共9块）
num_chunks = 3

# 计算每块的经纬度范围
lat_step = (north - south) / num_chunks
lon_step = (east - west) / num_chunks

last_amenities_filtered_chunk = None
last_population_filtered_chunk = None

# 逐块读取和保存数据
for i in tqdm(range(num_chunks), desc="Latitude chunks"):
    for j in range(num_chunks):
        # 计算每块的边界
        north_bound = north - i * lat_step
        south_bound = north_bound - lat_step
        west_bound = west + j * lon_step
        east_bound = west_bound + lon_step

        # 使用 OSMNX 读取特定范围内的设施和人口数据
        bbox = (south_bound, north_bound, west_bound, east_bound)

        # 读取设施数据
        amenities = ox.geometries_from_bbox(*bbox, tags={'amenity': True})
        useful_amenity_columns = ['amenity', 'name', 'geometry']
        amenities_filtered_chunk = amenities[useful_amenity_columns]

        # 读取人口数据
        population = ox.geometries_from_bbox(*bbox, tags={'population': True})
        useful_population_columns = ['population', 'geometry']
        population_filtered_chunk = population[useful_population_columns]

        # 保存每个块为 GeoJSON 文件
        amenities_filename = f'filtered_amenities_part_{i + 1}_{j + 1}.geojson'
        population_filename = f'filtered_population_part_{i + 1}_{j + 1}.geojson'
        amenities_filtered_chunk.to_file(amenities_filename, driver='GeoJSON')
        population_filtered_chunk.to_file(population_filename, driver='GeoJSON')

        # 打印保存信息
        print(f"Saved amenities chunk {i + 1}_{j + 1} to {amenities_filename}")
        print(f"Saved population chunk {i + 1}_{j + 1} to {population_filename}")
        print(amenities_filtered_chunk.head())
        print(population_filtered_chunk.head())

        # 释放内存
        del amenities, population
        gc.collect()

        # 只保留最后一个块的数据
        last_amenities_filtered_chunk = amenities_filtered_chunk
        last_population_filtered_chunk = population_filtered_chunk

# 最后一个块的数据在这里可用
print("Last amenities chunk:")
print(last_amenities_filtered_chunk.head())

print("Last population chunk:")
print(last_population_filtered_chunk.head())
```

```python
import osmnx as ox
import geopandas as gpd
from tqdm import tqdm
import pandas as pd

# Bounding box for New York City
north, south, east, west = 40.917577, 40.477399, -73.700272, -74.259090
bbox = (south, north, west, east)

# Extracting and saving building feature in three parts to manage memory usage
bbox_splits = [
    (south, (south + north) / 2, west, (west + east) / 2),
    (south, (south + north) / 2, (west + east) / 2, east),
    ((south + north) / 2, north, west, east)
]

building_filename = 'filtered_building_part_{}.geojson'

for i, split_bbox in enumerate(tqdm(bbox_splits, desc="Processing building data in parts")):
    building = ox.geometries_from_bbox(*split_bbox, tags={'building': True})
    building_filtered = building[['geometry', 'building']].dropna(subset=['building'])
    filename = building_filename.format(i + 1)
    building_filtered.to_file(filename, driver='GeoJSON')
    print(f"Saved building data part {i + 1} to {filename}")
    del building, building_filtered

# Print message indicating completion
print("Building data has been saved successfully.")
```
