```python
# 加载数据集
df = pd.read_csv('New_York_City_Population_By_Community_Districts_20240927.csv')

# 配置Google Maps客户端
gmaps = googlemaps.Client(key='AIzaSyCoO2HULPgeBZKvnxjtLWLn9mRdc7Kotb4')  # 替换为你的Google API密钥

# 定义地理编码函数
def geocode_address(address):
    try:
        geocode_result = gmaps.geocode(address + ', New York, NY')
        if geocode_result:
            latitude = geocode_result[0]['geometry']['location']['lat']
            longitude = geocode_result[0]['geometry']['location']['lng']
            return latitude, longitude
        else:
            return None, None
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
        return None, None

# 应用地理编码
df['Latitude'], df['Longitude'] = zip(*df['CD Name'].apply(geocode_address))

# 保存结果
df.to_csv('NYC_Population_with_Google_Coordinates.csv', index=False)
```
