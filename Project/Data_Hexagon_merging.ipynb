# Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime
```

```python

df = pd.read_csv('311_Noise_Complaints.csv')
```

```python
# Converteer alle kolomnamen naar kleine letters en vervang spaties door underscores
df.columns = df.columns.str.lower().str.replace(' ', '_')

# Convert the 'created_date' column to datetime
df['created_date'] = pd.to_datetime(df['created_date'], errors='coerce')

# Filter data for entries from 2020 onwards
df = df[df['created_date'] >= '2020-01-01']
```

```python
# Selecting relevant columns
columns_to_keep = ['location', 'complaint_type','descriptor']
df_filtered = df[columns_to_keep]
```

```python
# Check for missing data
print(df_filtered.isnull().sum())

# Drop rows with missing 'specified columns'
df_filtered.dropna(subset=['location', 'complaint_type','descriptor'], inplace=True)

df_filtered.head()

# Unieke waarden in de kolom 'complaint_type'
complaint_type_unique = df_filtered['complaint_type'].unique()

# Unieke waarden in de kolom 'descriptor'
descriptor_unique = df_filtered['descriptor'].unique()

print("Values in 'complaint_type':", complaint_type_unique)
print("Values in 'descriptor':", descriptor_unique)
```

```python
# Aantal voorkomen per unieke waarde in 'complaint_type', gesorteerd
complaint_type_counts = df_filtered['complaint_type'].value_counts().sort_index()

# Aantal voorkomen per unieke waarde in 'descriptor', gesorteerd
descriptor_counts = df_filtered['descriptor'].value_counts().sort_index()

#print("Amount per 'complaint_type':\n", complaint_type_counts)
print("\nAmount per 'descriptor':\n", descriptor_counts)
```

```python
# Voeg de kolommen 'complaint_type' en 'descriptor' samen in een nieuwe kolom met de gewenste naam
df_filtered['combined complaint type descriptor'] = df_filtered['complaint_type'] + ' - ' + df_filtered['descriptor']

# Controleer de eerste paar rijen van de dataset om te zien of de kolom correct is toegevoegd
print(df_filtered.head())
```

```python
# Tel het aantal voorkomen per unieke waarde in de nieuwe kolom
combined_counts = df_filtered['combined complaint type descriptor'].value_counts()

# Print de unieke waarden en hun aantal voorkomen
print(combined_counts)
```

```python
# Bereken het aantal voorkomen per unieke waarde in de kolom
combined_counts = df_filtered['combined complaint type descriptor'].value_counts()

# Maak een lijst van de waarden die vaker voorkomen dan (of gelijk aan) 200 keer
values_to_keep = combined_counts[combined_counts >= 200].index

# Maak een nieuwe DataFrame met alleen de rijen die voldoen aan de count-voorwaarde
df_filtered_2 = df_filtered[df_filtered['combined complaint type descriptor'].isin(values_to_keep)]

# Controleer de eerste paar rijen van de nieuwe DataFrame
#print(df_filtered_2.head())


combined_counts_2 = df_filtered_2['combined complaint type descriptor'].value_counts()
print(combined_counts_2)
```

```python
#STEP 1 Extract Latitude and Longitude from the ‘location’ column

# # Check the data type of the 'location' column
# print(df_filtered['location'].head())

# # If the location values are strings, convert them to tuples
# df_filtered['location'] = df_filtered['location'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# # Now extract latitude and longitude
# df_filtered[['latitude', 'longitude']] = pd.DataFrame(df_filtered['location'].tolist(), index=df_filtered.index)

# # Check the first few rows to confirm the split
# print(df_filtered[['latitude', 'longitude', 'combined complaint type descriptor']].head())


# Ensure proper handling and avoid SettingWithCopyWarning
df_filtered_2.loc[:, 'latitude'] = df_filtered_2['location'].apply(lambda x: x[0] if isinstance(x, tuple) else eval(x)[0])
df_filtered_2.loc[:, 'longitude'] = df_filtered_2['location'].apply(lambda x: x[1] if isinstance(x, tuple) else eval(x)[1])

# Check the first few rows to confirm the split
print(df_filtered_2[['latitude', 'longitude', 'combined complaint type descriptor']].head())
```

```python
# STEP 2 Hexagonal Binning using H3
#pip install h3

import h3

print(h3.__version__)

# # Define the resolution of the hexagons (higher number = smaller hexagons)
# hex_resolution = 8  # Adjust this number based on how small you want your hexagons

# # Create a new column 'hex_id' with the hexagon ID for each (latitude, longitude) pair
# df_filtered_2['hex_id'] = df_filtered_2.apply(lambda row: h3.geo_to_h3(row['latitude'], row['longitude'], hex_resolution), axis=1)

# # Check the first few rows to see the new hex_id column
# print(df_filtered_2[['hex_id', 'combined complaint type descriptor']].head())

import h3
# Resolution of the hexagons (higher number = smaller hexagons)
hex_resolution = 12  # Adjust this number based on how small you want your hexagons

# New column 'hex_id' with the hexagon ID for each (latitude, longitude) pair
df_filtered_2['hex_id'] = df_filtered_2.apply(lambda row: h3.latlng_to_cell(row['latitude'], row['longitude'], hex_resolution), axis=1)

# Check the first few rows to see the new hex_id column
#print(df_filtered_2[['hex_id', 'combined complaint type descriptor']].head())

# Tellen van het aantal unieke hexagons in de dataset
#aantal_hexagons = dominant_complaints['hex_id'].nunique()

# Print het resultaat
#print(f"Het aantal unieke hexagons in de dataset is: {aantal_hexagons}")
```

```python
# STEP 3  Aggregate Complaints by Hexagon

# Group by hexagon ID and 'combined complaint type descriptor' and count complaints in each hexagon
hex_complaint_counts = df_filtered_2.groupby(['hex_id', 'combined complaint type descriptor']).size().reset_index(name='complaint_count')

# Check the result
print(hex_complaint_counts.head())
```

```python
# VISUALS (FOLIUM) STEP 1 - Code om het Dominante Klachtentype per Hexagoon te Bepalen

import pandas as pd

# Stap 1: Bepaal het dominante klachtentype per hexagoon
# Groepeer per 'hex_id' en 'combined complaint type descriptor' om het aantal klachten per type te tellen
complaint_dominance = df_filtered_2.groupby(['hex_id', 'combined complaint type descriptor']).size().reset_index(name='complaint_count')

# Vind het klachtentype met het hoogste aantal per hexagoon
dominant_complaints = complaint_dominance.loc[complaint_dominance.groupby('hex_id')['complaint_count'].idxmax()]

# Stap 2: Maak een nieuwe kolom met de centroid (lat/long) van elke hexagoon
import h3

# Gebruik de correcte functie uit de h3-bibliotheek
dominant_complaints['latitude'] = dominant_complaints['hex_id'].apply(lambda x: h3.cell_to_latlng(x)[0])
dominant_complaints['longitude'] = dominant_complaints['hex_id'].apply(lambda x: h3.cell_to_latlng(x)[1])

# Controleer het resultaat
print(dominant_complaints.head())

```
